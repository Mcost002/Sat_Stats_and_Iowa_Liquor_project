##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def Std_dev(key):
    means = sum(key)/len(key)
    f = [(x-means)**2 for x in key]
    return (sum(f)/len(f))**.5
## we compute standard dev by hand. we create a func, create the math for mean and then in a list comp calculate
##the top half of the equation and pass the rest of it through return
```

```python
print "Standard Dev for Rate: " + str(Std_dev(Header_dict['Rate']))
print "Standard Dev for Verbal: " + str(Std_dev(Header_dict['Verbal']))
print "Standard Dev for Math: " + str(Std_dev(Header_dict['Math']))
```
